Récupérer via crawling la liste des 256 top contributors sur cette page https://gist.github.com/paulmillr/2657075
En utilisant l'API github (https://developer.github.com/v3/) récupérer pour chacun de ces users le nombre moyens de stars des repositories qui leur appartiennent. Pour finir, classer ces 256 contributors par leur note moyenne.﻿

Comme l'API github dispose de restrictions d'accès (limitation du nombre de requêtes), vous aurez besoin de vous authentifier via un token: https://help.github.com/en/articles/creating-a-personal-access-token-for-the-command-line.

```python
import requests
import urllib.request
from bs4 import BeautifulSoup
import time

import pandas as pd
```

```python
url = "https://gist.github.com/paulmillr/2657075"
```

Récupération de la liste des 256 tops contributors :

```python
def git_crawling(url):
 
    page = requests.get(url, verify=True, timeout = 5)
    html = page.text
    soup = BeautifulSoup(html, "html.parser")
    divs = soup.find('table').find('tbody').find_all('tr')
    
    Users_list=[]
    for index, line in enumerate(divs):
        name = line.find('td').text.split(" ")[0]
        Users_list.append(name)
    
    return Users_list
```

```python
git_crawling(url)
```

Récupération du nombre moyens de stars sur les repos de ces 256 contributors.

```python
def get_mean_star_repo(user):
    
    url = f'https://api.github.com/users/{user}/repos'
    headers = {'Authorization': 'Token XXXXXXXXXXXXXXXXXXXXXXXXXXXXXX'} 
    
    #to authenticate with a token
    res = requests.get(url, headers=headers)
    repositories = res.json()
    
    #compute the average
    total_star = 0
    for repo in repositories:
        total_star += repo['stargazers_count']
        
    #case of null 
    if len(repository) == 0:
        return 0
    
    average_star = total_star / len(repository)

    return average_star
```

Obtention du résultat final :

```python
def bests_users(url):
    list = []
    for user in git_crawling(url)[0:50]: #limited to 50 for that example
        info_user = {'user': user, 'mean_star' : get_mean_star_repo(user) }
        list.append(info_user)
    return list
```

```python
Final_result = pd.DataFrame(bests_users(url))
```
